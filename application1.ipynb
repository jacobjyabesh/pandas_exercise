```python
import pandas as pd
import numpy as np
```

```python
city = ['CHENNAI', 'MUMBAI', 'HYDERABAD']
companies=['ABC limited', 'XYZ limited']
```

```python
emp = pd.DataFrame({'NAME': ['name'+str(i) for i in range(1,16)],
                    'CITY': [city[np.random.randint(0,3)] for i in range(1,16)],
                    'STREET': ['street'+str(i) for i in range(1,16)]})
```

```python
company = pd.DataFrame({'NAME': [companies[i] for i in range(len(companies))],
                        'CITY': [city[np.random.randint(0,3)] for i in range(len(companies))]})
```

```python
works = pd.DataFrame({'EMPNAME': ['name'+str(i) for i in range(1,16)],
                      'COMPNAME': [companies[np.random.randint(0,2)] for i in range(1, 16)],
                      'SALARY': np.random.randint(10000,100000,15)})
```

```python
manages = pd.DataFrame({'EMPNAME': ['name'+str(i) for i in range(1,16)],
                        'MANAGERNAME': ['name'+str(np.random.randint(0,3)) for i in range(1,16)]})
```

```python
#1. Find names of employee who works for 'ABC limited'
works[works['COMPNAME']=='ABC limited']['EMPNAME']
```

```python
#2. Find the name, street and location of employees who work for 'ABC limited' and earm more than 40000
emp[emp['NAME'].isin(works[(works['COMPNAME']=='ABC limited') & (works['SALARY']>40000)]['EMPNAME'])]
```

```python
#3. find the names of emaployess whose location and their compnay location is same
emp.merge(works.merge(company, left_on='COMPNAME', right_on='NAME', how='inner'), left_on=['NAME', 'CITY'], right_on=['EMPNAME', 'CITY'], how='inner')
```

```python
#4. Find the name of all employees who earn more than name3
works[works['SALARY'] > works[works['EMPNAME']=='name3']['SALARY'].values[0]]
```

```python
#5. Find names of employee who don't work for 'ABC Limited'
works[~(works['COMPNAME']=='ABC limited')]['EMPNAME']
```
