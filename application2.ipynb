```python
import pandas as pd
import numpy as np
```

```python
gender = ['M', 'F']
```

```python
emp = pd.DataFrame({'ID':[i for i in range(1,16)],
                    'NAME': ['name'+str(i) for i in range(1,16)],
                    'GENDER': [gender[np.random.randint(0,2)] for i in range(1,16)],
                    'DOB': pd.date_range(start='01/01/1960', end='31/12/1990', periods=15, normalize=True).strftime('%Y-%m-%d'),
                    'SALARY': np.random.randint(10000,160000, 15),
                    'DEPTID': np.random.randint(1,4, 15),
                    'DOJ': pd.date_range(start='01/01/2012', end='12/12/2019', periods=15).strftime('%Y-%m-%d')})
```

```python
dept = pd.DataFrame({'ID': [1,2,3],
                     'NAME': ['IT', 'ACCOUNTS', 'SALES']})
```

```python
project = pd.DataFrame({'ID': [1,2],
                        'NAME': ['ML', 'AI'],
                        'DEPTID': [1,3]})
```

```python
workfor = pd.DataFrame({'EMPNO': np.random.randint(1,16,20),
                       'PROJNO': np.random.randint(1,3,20)}
                       )
```

```python
emp
```

```python
dept
```

```python
project
```

```python
workfor
```

```python
#1. List the name of all the employes whose salary is less than the average salary of the company
emp[emp['SALARY'] < emp[['SALARY']].agg(['mean'])['SALARY'].values[0]]
```

```python
#2. List the Dept name and number of employees working in each dept
emp.merge(dept, left_on=['DEPTID'], right_on=['ID'], suffixes=['_E', '_D'])[['NAME_D']].groupby('NAME_D').size()
```

```python
#3. List the details of employee who work for a project conrolled by SALES department
emp[emp['ID'].isin(workfor[workfor['PROJNO'].isin(project[project['DEPTID']==dept[dept['NAME']=='SALES']['ID'].values[0]]['ID'])]['EMPNO'])]
```

```python
#4. List names of employee whose name is between 100000 and 150000
emp[emp['SALARY'].between(100000, 150000)]['NAME']
```

```python
#5. list the details of employee if he works in more than a project
emp_proj = workfor.groupby('EMPNO').agg(['count'])
emp[emp['ID'].isin(emp_proj[emp_proj[('PROJNO', 'count')]>1].index)]
```

<h1>New Application</h1>

```python
book = pd.DataFrame({'TITLE': ['ML', 'AI', 'MATHS', 'Python', 'Data structures', 'Computer Networks'],
                     'AUTHOR': ['JOHN', 'JOHN', 'PAUL', 'MARK', 'MATHEW', 'TITUS'],
                     'PUBLISHER': ['PEARSON', 'XYZ', 'ABC', 'ABC', 'PEARSON', 'JP Publications'],
                     'YEAR': [1990, 2008, 2001, 2019, 2015, 2008],
                     'PRICE': [100, 80, 20, 90, 300, 95],
                     'COUNTRY': ['USA', 'USA', 'UK', 'INDIA', 'CANADA', 'CHINA']})
```

```python
book
```

```python
#1. Display details about books except those published in 2008 and 2019
book[~book['YEAR'].isin([2008, 2019])]
```

```python
#2. Display details about books those published in 2008 and 1990
book[book['YEAR'].isin([2008, 1990])]
```

```python
#3. List the title of books written by author not living in INDIA
book[book['COUNTRY']!='INDIA']['TITLE']
```

```python
#4. list the books whose price is greater than the books writeen in the year 2008
book[book['PRICE']>book[book['YEAR']==2008]['PRICE'].max()]
```

```python
#5. Find the title of book having price in the range 90 and 150
book[book['PRICE'].between(90, 150)]['TITLE']
```

```python
#6 FInd the name of authour who wrote more than a book
author_book = book.groupby(by=['AUTHOR']).agg(['count'])
author_book[author_book[('TITLE', 'count')]>1].index.values
```

```python
customer = pd.DataFrame({'NAME': ['A', 'B', 'C', 'D', 'JACKSON'],
                         'STREET': ['S1', 'S2', 'S3', 'S4', 'S5'],
                         'CITY': ['CHENNAI', 'NELLAI', 'TRICHY', 'CHENNAI', 'CHENNAI']})
```

```python
account = pd.DataFrame({'ACCNO': [101, 102, 103],
                        'BALANCE': [0, 20000, 100000],
                        'BNAME': ['X', 'Y', 'X']})
```

```python
loan = pd.DataFrame({'LOANNO': [201, 202, 203],
                     'AMOUNT': [0, 210000, 12000],
                     'BNAME': ['Z', 'Y', 'Z']})
```

```python
depositor = pd.DataFrame({'CNAME': ['A', 'B', 'C'],
                          'ACCNO': [101, 102, 103]})
```

```python
borrower = pd.DataFrame({'CNAME': ['C', 'D', 'JACKSON'],
                         'LOANNO': [201, 202, 203]})
```

```python
#1. Find the name of customer who has loan over 100000
borrower[borrower['LOANNO'].isin(loan[loan['AMOUNT']>100000]['LOANNO'])]['CNAME']
```

```python
#2. find the name of customer who has more than 1000 in their account
depositor[depositor['ACCNO'].isin(account[account['BALANCE']>1000]['ACCNO'])]['CNAME']
```

```python
#3. find the name of customer who has loan and deposit
depositor[['CNAME']].merge(borrower[['CNAME']])
```

```python
#4. Find the name of customer who has loan or deposit or both
pd.concat([depositor[['CNAME']], borrower[['CNAME']]], ignore_index=True).drop_duplicates()
```

```python
#5. Find the name of customer who has only deposit
depositor[depositor['CNAME'].isin(borrower['CNAME'])==False]
```

```python
#6. Customer who has only loan
borrower[borrower['CNAME'].isin(depositor['CNAME'])==False]
```

```python
#7. Find the name of customers who has loan at branch 'Z'
borrower[loan.merge(borrower)['BNAME']=='Z']
```
